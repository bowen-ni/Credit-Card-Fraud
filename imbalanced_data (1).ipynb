- Handling Imbalanced Datasets

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy='minority', random_state=42)

X_smote, y_smote = sm.fit_resample(df.drop('Class', axis=1), df['Class'])
smote_df = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote)], axis=1)
X_smote.shape

smote_df.Class.value_counts()
```

```python
datasets = {

    'SMOTE': (X_smote, y_smote)
}
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

results = []
```

```python
for sampling_name, (X_data, y_data) in datasets.items():
    # Split into train and test sets (30% test)
    X_train, X_test, y_train, y_test = train_test_split(
        X_data, y_data, test_size=0.3, random_state=42, stratify=y_data
    )

    for model_name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        # Predict on test set
        y_pred = model.predict(X_test)

        # Calculate metrics
        acc = accuracy_score(y_test, y_pred)
        cm = confusion_matrix(y_test, y_pred)
        cr = classification_report(y_test, y_pred, output_dict=True)

        # Store summary of results
        results.append({
            'Sampling Method': sampling_name,
            'Model': model_name,
            'Accuracy': acc,
            'Precision (Fraud)': cr['1']['precision'],
            'Recall (Fraud)': cr['1']['recall'],
            'F1-score (Fraud)': cr['1']['f1-score']
        })

        # Print detailed reports
        print(f"\n=== {sampling_name} - {model_name} ===")
        print(f"Accuracy: {acc:.4f}")
        print("Confusion Matrix:")
        print(cm)
        print("Classification Report:")
        print(classification_report(y_test, y_pred))

        # Plot confusion matrix heatmap
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusion Matrix: {sampling_name} - {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
```

```python
results_df = pd.DataFrame(results)
```

```python
# SMOTE-only results
smote_results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'F1 Score': [0.84, 0.91, 0.93],
    'Accuracy': [0.985, 0.992, 0.997]
})

# Plotting F1 Score and Accuracy
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# F1 Score plot
ax[0].bar(smote_results['Model'], smote_results['F1 Score'], color='skyblue')
ax[0].set_title('F1 Score for SMOTE')
ax[0].set_ylabel('F1 Score')
ax[0].set_ylim(0, 1.05)

# Accuracy plot
ax[1].bar(smote_results['Model'], smote_results['Accuracy'], color='lightgreen')
ax[1].set_title('Accuracy for SMOTE')
ax[1].set_ylabel('Accuracy')
ax[1].set_ylim(0, 1.05)

plt.suptitle('Model Performance with SMOTE Sampling')
plt.tight_layout()
plt.show()






```
